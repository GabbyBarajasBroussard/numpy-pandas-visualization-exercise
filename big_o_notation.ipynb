```python
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

x1= list(range(0,25))
y1= [((0*n)+1) for n in x1]
plt.plot(x1, y1, c='teal', label= '$O(1)$')


x2= range(1,6)
y2= [math.log(n) for n in x2]
plt.plot(x2, y2, c='fuchsia', label= 'O $log n$')

x3= range(1,6)
y3= [n*1 for n in x3]
plt.plot(x3, y3, c='lightblue', label= '$O(n)$')

x4= range(1,6)
y4= [(n * math.log(n)) for n in x4]
plt.plot(x4, y4, c='darkmagenta', label='O n $\log n$')

x5= range(1,6)
y5= [(n**2) for n in x5]
plt.plot(x5, y5, c='orchid', label = '$O(n^2)$')

x6= range(1,6)
y6= [(2**n) for n in x6]
plt.plot(x6, y6, c='turquoise', label='$O(2^n)$')

x7=[1,2,3]
y7= [math.factorial(n) for n in x7]
plt.plot(x7, y7, c='cornflowerblue', label='$O(n!)$')

x8= range(1,6)
y8= [(n**n) for n in x8]
plt.plot(x8, y8, c='mediumblue', label='$O n^n$')

plt.ylim(-1, 16)
plt.xlim(1, 4)

plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(16, 9)) #setting the figure size
plt.title('Big O Notation')
x= list(range(1,10))

y1= [((0*n)+1) for n in x]
y2= [math.log(n) for n in x]
y3= [n*1 for n in x]
y4= [(n * math.log(n)) for n in x]
y5= [(n**2) for n in x]
y6= [(2**n) for n in x]
y7= [math.factorial(n) for n in x]
y8= [(n**n) for n in x]

plt.plot(x, y1, c='teal', label= '$O(1)$')
plt.plot(x, y2, c='fuchsia', label= 'O $log n$')
plt.plot(x, y3, c='lightblue', label= '$O(n)$')
plt.plot(x, y4, c='darkmagenta', label='O n $\log n$')
plt.plot(x, y5, c='orchid', label = '$O(n^2)$')
plt.plot(x, y6, c='turquoise', label='$O(2^n)$')
plt.plot(x, y7, c='cornflowerblue', label='$O(n!)$')
plt.plot(x, y8, c='mediumblue', label='$O n^n$')

plt.ylim(1, 11)
plt.xlim(1, 11)
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.legend(loc='upper left')
```
